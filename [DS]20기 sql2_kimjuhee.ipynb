문제 7.

위도, 경도 정보를 이용하여 미왕빌딩에서 셀프주유가 가능하고 24시간이면서 편의점이 있는 가장 가까운 주유소 10개를 휘 발유 가격이 가장 저렴한 순으로 정렬하여 조회하세요.

- 주유소 아이디, 주유소 브랜드명, 주유소 매장명, 주소, 휘발유 가격, 부가정보 (셀프, 24시간, 편의점 여부), 미왕빌딩으로 부터의 거리 (km)


제출 7.
- 쿼리, 결과

```python
cursor.execute('''
    SELECT a.id, a.brand, a.name, a.address, a.gasoline, a.self, a.24_hours, a.convenience_store, a.distance
    FROM (
        SELECT *, ST_DISTANCE_Sphere(POINT(lng, lat), POINT(127.0292924, 37.4955525)) * 0.001 AS distance 
        FROM GAS_STATION 
        WHERE self=1 AND 24_hours=1 AND convenience_store=1 
        ORDER BY distance 
        LIMIT 10
    ) a
    ORDER BY  a.gasoline
''')
cursor.fetchall()
```

```python
cursor.execute('''
SELECT a.id, GB.NAME, a.name, a.address, a.gasoline, a.self, a.24_hours, a.convenience_store, a.distance
    FROM (
        SELECT *, ST_DISTANCE_Sphere(POINT(lng, lat), POINT(127.0292924, 37.4955525)) * 0.001 AS distance 
        FROM GAS_STATION 
        WHERE self=1 AND 24_hours=1 AND convenience_store=1 
        ORDER BY distance 
        LIMIT 10
    ) a left join GAS_BRAND GB ON a.BRAND = GB.ID
    ORDER BY  a.gasoline
''')
cursor.fetchall()
```

```python
cursor.execute('''select GS.gu, GB.name, GS.gasoline
from (select gu, brand, avg(gasoline) AS gasoline
               from GAS_STATION 
               group by gu, brand
               order by avg(gasoline)) GS left join GAS_BRAND GB ON GS.BRAND = GB.ID''')
cursor.fetchall()
```
